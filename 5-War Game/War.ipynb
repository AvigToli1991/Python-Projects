```python
import random
from Deck import Deck
from Player import player
```

```python
from IPython.display import clear_output

def game() : 

    replay = True
    

    player1 = player(input("First player enter your name please: "))
    player2 = player(input("Second player enter your name please: "))
    
    while replay:
        
        main_deck = Deck(all=True)

        for i in range(3):
            main_deck.suffle()

        player1.create_main(main_deck)
        player2.create_main(main_deck)
        
        at_war = True
        
        while at_war: 

            if player1.is_empty():
                print(f"{player1.name} is out of cards!")
                print(f"The winner is {player2.name}")
                at_war = False
                break
            elif player2.is_empty():
                print(f"{player2.name} is out of cards!")
                print(f"The winner is {player1.name}")
                at_war = False
                break
            else:
                war_deck = []
                print("\n")
                war_deck.append(player1.draw_one()) 
                print(war_deck[-1])
                print("\n")
                war_deck.append(player2.draw_one())
                print(war_deck[-1])
                print("\n")

                while war_deck[-1].value == war_deck[-2].value:
                    
                    print("Its a War!!!\n")
                    
                    len_deck_player1 = player1.get_len_of_deck() 
                    len_deck_player2 = player2.get_len_of_deck()
                    
                    if len_deck_player1 < 4:
                        print(f"{player1.name} have no cards enough to make war")
                        print(f"The winner is {player2.name}!")
                        at_war = False
                        break
                    elif len_deck_player2 < 4:
                        print(f"{player1.name} have no cards enough to make war")
                        print(f"The winner is {player2.name}!")
                        at_war = False
                        break
                    else:
                        for i in range(3):
                                war_deck.append(player1.draw_one())
                                war_deck.append(player2.draw_one())
                        print('\n')
                        war_deck.append(player1.draw_one())
                        print(war_deck[-1],'\n')
                        
                        war_deck.append(player2.draw_one())
                        print(war_deck[-1],'\n')
                        
                else:
                    if war_deck[-2].value > war_deck[-1].value:
                        print(f"{player1.name} take this war!")
                        player1.add_cards(war_deck)
                    else: 
                        print(f"{player2.name} take this war!")
                        player2.add_cards(war_deck)
            
                    
        gameOn = input("whuld you like to play again?(y/n)")                     
        while (gameOn not in ['y','n']):
            gameOn = input("whuld you like to play again?(y/n)") 

        if gameOn == 'n':
            replay = False
            print("Thanks for playing")
        else:
            clear_output()

        
```
